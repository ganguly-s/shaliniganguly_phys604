```python
import numpy as np
n = 1
a = 1
b = 1
c = 10**(-n)
d = np.sqrt(b*b - 4*a*c)
r1 = (-b + d)/(2*a)
r2 = (-b - d)/(2*a)
r3 = (-2*c)/(b + d)
r4 = (-2*c)/(b - d)
print('First set:', (r1,r2))
print('Second set:', (r3,r4))
```

```python
for n in range(1,50):
    c = 10**(-n)
    d = np.sqrt(b*b - 4*a*c)
    r1 = (-b + d)/(2*a)
    r2 = (-b - d)/(2*a)
    r3 = (-2*c)/(b + d)
    r4 = (-2*c)/(b - d)
    print('n = ', n)
    print('First set:', (r1,r2))
    print('Second set:', (r3,r4))
```

```python
#Machine precision suggests correct calculation only upto n=16

n = 1
a = 1
b = 1
c = 10**(-n)
d = np.sqrt(b*b - 4*a*c)
r2 = (-b - d)/(2*a)
r3 = (-2*c)/(b + d)
print('Most precise solution: ', (r2,r3))
```
