```python
import numpy as np
import time

#Iteration method

val = 0
val1 = 0
val2 = 1
value = []
f = open('iteration.txt','w')
def fibbo1(n):
    val = 0
    val1 = 0
    val2 = 1
    value = []
    ti = time.time()
    for i in range(n-2):
        val = val1 + val2
        val1 = val2
        val2 = val
    value.append(n)
    value.append(time.time()-ti)
    return value
for i in range(100,10000,1000):
    v = fibbo1(i)
    str1 = str(v[0])
    str2 = str(v[1])
    f.write(str1+' '+str2+'\n')
f.close()
```

```python
n1 = []
t1 = []
with open('iteration.txt') as f:
    for line in f:
        str1=line.split()
        num=[float(x) for x in str1]
        n1.append(num[0])
        t1.append(num[1])
from matplotlib import pyplot as plt
plt.plot(n1,t1)
plt.xscale('log')
plt.xlabel('n')
plt.ylabel('t')
plt.show()
```

```python
#Tabulation method

def fibbo2(n):      # fast, no stack limit
    table=[0]*(n+1)
    table[1] = 1
    
    for i in range(2,n+1):
        table[i]=table[i-1]+table[i-2]
    return table[n]
```

```python
f = open('tabulation.txt','w')
for i in range(100,10000,2000):
    ti = time.time()
    v = fibbo2(i)
    str1 = str(i)
    str2 = str(time.time()-ti)
    f.write(str1+' '+str2+'\n')
f.close()
```

```python
#Matrix method

def arrpow(arr, n): # calculate 2x2 array to the nth power.
    yarr=arr
    if n<1:
        print('n needs to be larger than 1')
    if n==1:
        return arr
    yarr = arrpow(arr, n//2)
    yarr = [[yarr[0][0]*yarr[0][0]+yarr[0][1]*yarr[1][0],yarr[0][0]*yarr[0][1]+yarr[0][1]*yarr[1][1]], # array multiply itself
            [yarr[1][0]*yarr[0][0]+yarr[1][1]*yarr[1][0],yarr[1][0]*yarr[0][1]+yarr[1][1]*yarr[1][1]]]
    if n%2:   # if n is an odd number, multiple another arr
        yarr=[[yarr[0][0]*arr[0][0]+yarr[0][1]*arr[1][0],yarr[0][0]*arr[0][1]+yarr[0][1]*arr[1][1]],
            [yarr[1][0]*arr[0][0]+yarr[1][1]*arr[1][0],yarr[1][0]*arr[0][1]+yarr[1][1]*arr[1][1]]]
    return yarr

def fibbo3(n):   # the fastest method
    arr= [[1,1],[1,0]]
    f=arrpow(arr,n-1)[0][0]  #0 0 component is Fn#
    return f
```

```python
f = open('matrix.txt','w')
for i in range(100,10000,2000):
    ti = time.time()
    v = fibbo3(i)
    str1 = str(i)
    str2 = str(time.time()-ti)
    f.write(str1+' '+str2+'\n')
f.close()
```
