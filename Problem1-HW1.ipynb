```python
import numpy as np
from matplotlib import pyplot as plt

def fib3(n, table): 
    if n == 0 or n==1:
        table[n]=n
    if table[n] is None:
        table[n] = fib3(n-1, table) + fib3(n-2, table)
    return table[n]

x = np.arange(2,20,1)
y = []
for i in range(2,20):
    n = i
    table = [None]*(n+1)
    num = fib3(n, table)
    n = i - 1
    table = [None]*(n+1)
    den = fib3(n, table)
    y.append(num/den)

plt.plot(x,y)
plt.xlabel('n')
plt.ylabel(r'$F_n/F_{n-1}$')
plt.show()
```
