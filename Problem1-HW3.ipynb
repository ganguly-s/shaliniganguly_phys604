```python
import numpy as np
from matplotlib import pyplot as plt

x0=np.pi/4.
dx=0
def derifw(y,x,dx):
    return (y(x+dx)-y(x))/dx
def logerr(fn,fa):
    return np.log10(abs(fn-fa))

x = []
y = []
for i in range(16,0,-1):
    dx = 10**(-i)
    y.append(logerr(derifw(np.cos,x0,dx),-np.sin(x0)))
    x.append(np.log10(dx))

#plot shows error is proportional to dx upto dx=1e-8, below which error increases again
#as dx is reduced, error reduces linearly 
#error~dx^n, where n~1
#optimal dx value is dx=1e-8

plt.plot(x,y) 
plt.xlabel(r'$log_{10}(dx)$')
plt.ylabel(r'$log_{10}(error)$')
plt.xlim(min(x),max(x))
plt.ylim(min(y),max(y))
plt.show()    
```

```python
def dericd(y,x,dx):
    return (y(x+dx/2)-y(x-dx/2))/dx
x = []
y = []
for i in range(16,0,-1):
    dx = 10**(-i)
    y.append(logerr(dericd(np.cos,x0,dx),-np.sin(x0)))
    x.append(np.log10(dx))

#we can see from the plot that error~dx^n where n~2
#the error~O(dx^2), which is better than forward difference method where error~O(dx)

plt.plot(x,y) 
plt.xlabel(r'$log_{10}(dx)$')
plt.ylabel(r'$log_{10}(error)$')
plt.xlim(min(x),max(x))
plt.ylim(min(y),max(y))
plt.show()    
```
